##find factorial of given number

```python
num=int(input("enter a number: "))
fact=1
for i in range(1,num+1):
  fact=fact*i
print("factorial is",fact)
```

##fibonacci sequence

```python
a=0
b=1
c=int(input("enter a value of c: "))
print(a)
print(b)
for i in range(1,c+1):
  c=a+b
  a=b
  b=c
  print(c)
```

##find the sum of natural number

```python
a=int(input("enter a number: "))
if(a<0):
  print("enter postive number")
else:
  sum=0
  while(a>0):
    sum=sum+a
    a=a-1
print(sum)
```
