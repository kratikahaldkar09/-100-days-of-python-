##comparison operator

```python
a=28
b=8
print('equality:',a==b)
print('not equal:',a!=b)
print('greater than:',a>b)
print('less than:',a<b)
print('greater than or equal to:',a>=b)
print('less than:',a<=b)
```

##logical operators


```python
a=8
b=70
print(a or b)
print(a and b)
print(not a)
print(not b)
```

##identity operator

```python
a=100
b=99+1
print(a==b)
print(a is b)
print(id(a))
print(id(b))
```

##membership operators

```python
a=['python','java','cpp']
print('cpp' in a)
print('javaScript'in a)
print('C##'not in a)
```
